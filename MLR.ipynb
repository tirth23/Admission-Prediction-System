```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Admission_Predict_Ver1.1.csv')
print(df)
```

```python
df.isna().sum()
```

```python
print('GRE Score mean and median')
print(df['GRE Score'].mean())
print(df['GRE Score'].median())
print('TOEFL Score mean and median')
print(df['TOEFL Score'].mean())
print(df['TOEFL Score'].median())
print('University Rating mean and median')
print(df['University Rating'].mean())
print(df['University Rating'].median())
print('SOP mean and median')
print(df['SOP'].mean())
print(df['SOP'].median())
print('LOR mean and median')
print(df['LOR '].mean())
print(df['LOR '].median())
print('CGPA mean and median')
print(df['CGPA'].mean())
print(df['CGPA'].median())
print('Research mean and median')
print(df['Research'].mean())
print(df['Research'].median())
print('Chance of Admit mean and median')
print(df['Chance of Admit '].mean())
print(df['Chance of Admit '].median())
```

```python
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
print(x)
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x= sc.fit_transform(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 950) 
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
print(y_pred)
np.set_printoptions(precision=2)    
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
import sklearn.metrics as met
mse = met.mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = met.r2_score(y_test,y_pred)
print('Mean Squared Error: ',mse)
print('Root Mean Squared Error: ',rmse)
print('R-squared Score: ',r2)
```

```python
X_back = x[:,:]
print(X_back)
X_back = np.append(arr = np.ones((500,1)).astype(int), values = X_back, axis = 1)   
print(X_back)
```

```python
import statsmodels.api as sm
X_opt = np.array(X_back, dtype = float)
print('X_opt with all features')
print(X_opt)
sl = 0.05
numVars = len(X_opt[0])
for i in range(0,numVars):
    reg = sm.OLS(endog = y, exog = X_opt).fit()
    max_p_value = max(reg.pvalues).astype(float)
    if max_p_value > sl:
        for j in range(0, numVars-i):
            if(reg.pvalues[j].astype(float)==max_p_value):
                X_opt = np.delete(X_opt,j,1)
    print(reg.summary())
print('X_opt with most significant feature')
print(X_opt)
```

```python
x_tr,x_te,y_tr,y_te = train_test_split(X_opt,y,random_state = 950)
lr_opt = LinearRegression()
lr_opt.fit(x_tr,y_tr)
```

```python
y_pr = lr_opt.predict(x_te)
print(y_pr)
print(np.concatenate((y_pr.reshape(len(y_pr),1),y_te.reshape(len(y_te),1)),1)) 
```

```python
import sklearn.metrics as met
mse = met.mean_squared_error(y_te,y_pr)
rmse = np.sqrt(mse)
r2 = met.r2_score(y_te,y_pr)
print('Mean Squared Error: ',mse)
print('Root Mean Squared Error: ',rmse)
print('R-squared Score: ',r2)
```
