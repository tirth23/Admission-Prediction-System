```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset =pd.read_csv('Admission_Predict_Ver1.1.csv')
print(dataset)
```

```python
print('GRE Score mean and median')
print(dataset['GRE Score'].mean())
print(dataset['GRE Score'].median())
print('TOEFL Score mean and median')
print(dataset['TOEFL Score'].mean())
print(dataset['TOEFL Score'].median())
print('University Rating mean and median')
print(dataset['University Rating'].mean())
print(dataset['University Rating'].median())
print('SOP mean and median')
print(dataset['SOP'].mean())
print(dataset['SOP'].median())
print('LOR mean and median')
print(dataset['LOR '].mean())
print(dataset['LOR '].median())
print('CGPA mean and median')
print(dataset['CGPA'].mean())
print(dataset['CGPA'].median())
print('Research mean and median')
print(dataset['Research'].mean())
print(dataset['Research'].median())
```

```python
X=dataset.iloc[:,1:7].values
y=dataset.iloc[:,8].values
```

```python
from sklearn.preprocessing import StandardScaler
sc_X= StandardScaler()
X = sc_X.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X = pca.fit_transform(X)
p_var = pca.explained_variance_ratio_
print(p_var)
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.25,random_state=950)
```

```python
from sklearn.ensemble import RandomForestRegressor
classifier = RandomForestRegressor(n_estimators=100,random_state=200)
classifier.fit(X_train,y_train)
```

```python
Y_pred = classifier.predict(X_test)
print(Y_pred)
```

```python
errors = abs(Y_pred - y_test)
mape = 100 * (errors / y_test)
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
import sklearn.metrics as met
mse = met.mean_squared_error(y_test,Y_pred)
rmse = np.sqrt(mse)
r2 = met.r2_score(y_test,Y_pred)
print('Mean Squared Error: ',mse)
print('Root Mean Squared Error: ',rmse)
print('R-squared Score: ',r2)
```

```python
from mpl_toolkits import mplot3d
ax = plt.figure().add_subplot(projection = '3d')
ax.scatter(X_test[:,0],X_test[:,1],y_test,c = 'red')
ax.plot(X_test[:,0],X_test[:,1],Y_pred,c = 'blue')
```
